```python
import pandas as pd
```

```python
nc = pd.read_csv("311-noise-complaints2.csv")
nc
```

```python
mi = pd.read_csv("median-income-2021-raw.csv")
mi
```

```python
##filter large median census data to smaller dataset
midf = mi.filter(["NAME", "S1901_C01_012E"])
midf
```

```python
## rename columns 
## can put the same variable in to save the changes back into the variable
midf = midf.rename(columns={"NAME": "zip", "S1901_C01_012E": "median-income"})
midf
```

```python
midf = midf.drop(midf.index[0])
midf
```

```python
midf.info()
```

```python
## turn income into floating point numbers or integers before exporting
midf["median-income"] = pd.to_numeric(midf["median-income"], errors = "coerce")
midf.info()
```

```python
midf["zip"] = midf["zip"].str.replace("ZCTA5", "")
midf
```

```python
##export cleaned up version as a csv file
##add index mention so it doesn't re-add it again
midf.to_csv("median-income-2021.csv", index = False)
```
