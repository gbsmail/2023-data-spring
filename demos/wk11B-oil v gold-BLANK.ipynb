# Oil v. Gold Comparison

You want to compare the monthly closing price of gold v. oil over the past decade.

The easiest way to uncover a pattern will be to line graph the data

The data can be [found here](https://drive.google.com/file/d/122tdS2eTZEq9q7bbKEZMf6168Vi8biRh/view?usp=share_link).



 

## What is our approach?

```python
import pandas as pd
```

```python
df_raw = pd.read_csv("oil-v-gold.csv")
df_raw
```

```python
###check if the dates are lined up
df_raw.sample(20)
```

```python
df_raw.info()
```

## Fix it 
## drop duplicates and nan values
## convert objects into floating point numbers
## drop column names, provide sensical titles
## rename gold and oil columns

```python
## drop duplicate columns
df_raw.columns
```

```python
df = df_raw[['GC1 Comdty (Gold)','Unnamed: 1', 'Unnamed: 4' ]]
df
```

```python
## rename column headers

df.columns = ["closing_date", "px_gold", "px_oil"]
df
```

```python
## delete top row
## reset index so it still follows the counting convention of starting at 0
df.drop(0).reset_index(drop = True)
```

```python
df.info()
```

```python
## convert to floating point numbers 
## "COERCE" means turning non numbers into a NAN value 
df["px_gold"] = pd.to_numeric(df["px_gold"], errors = "coerce").copy()
df.info()
```

```python
df["px_oil"] = pd.to_numeric(df["px_oil"], errors = "coerce").copy()
df.info()
```

```python
## export to csv for a clean copy of data
df.to_csv("oil-v-gold-export.csv", index = False, encoding = "UTF8")
```

```python
df
```

```python
## creat quick line graphs of oil v gold. but it's not useful for us right now
df.plot(x="closing_date", y = ["px_gold", "px_oil"])
```

```python
## percent change calculation for gold using iat 

## fixed value
old_gold = df["px_gold"].iat[1]
old_gold
```

```python
old_oil = df["px_oil"].iat[1]
old_oil
```

```python
df["gold_pct_chg"] = \
((df["px_gold"] - old_gold) / old_gold) * 100
df
```

```python
df["oil_pct_chg"] = \
((df["px_oil"] - old_oil) / old_oil) * 100
df
```

```python
## basic line graph 
df.plot(x = "closing_date", y = ["oil_pct_chg", "gold_pct_chg"] )
```

```python
## add a title
df.plot(x = "closing_date", y = ["oil_pct_chg", "gold_pct_chg"], title = "Oil v. Gold Over Time" )
```

```python
## why is gold orange and oil blue? 
## can styleit! 
df.plot(x = "closing_date", 
        y = ["oil_pct_chg", "gold_pct_chg"], 
        title = "Oil v. Gold Over Time", 
        style = {"oil_pct_chg": "gray",
                 "gold_pct_chg": "gold"})
```
